```python
import pandas as pd
import os
import seaborn as sns
import warnings

# Suppress all warnings
warnings.filterwarnings('ignore')


```

Steps:


1.   Load the csv file with attributes
2.   Analyse te distribution of gender
3.   Analyse hair color distribution
4.   See how hair color and gender are related to understand data biases

```python
filename='Tiny_Portraits_Attributes.csv'

if not os.path.exists(filename):
    print('Error: filename does not exist!')
else:
    df=pd.read_csv(filename)

df.head()
```

```python
df["Gender"] = df['Gender'].astype('category')
sns.countplot(df, x='Gender')
```

```python
df1=df['Hair_Color']
df1.dropna(axis=0,inplace=True)
df1["Hair_Color"] = df['Hair_Color'].astype('category')
sns.countplot(df1, x= "Hair_Color")
```

```python
df2=df[['Gender','Hair_Color']]
df2.dropna(axis=0,inplace=True)
df2.hist(column='Hair_Color',by='Gender')
```

Analysis Summary:


*   Gender: Datset contains more females than males. Hence while training classifier use this information
*   Hair color: Classes are imbalanced. Bald and gray hair colors are very rare.
*   Hair color by itself depends on the gender. If we plot hair color of male and female we can observe that in males the dominant color is black. While in female brown and blonde. 
*  Bald and gray hair color in  females are very scarse
